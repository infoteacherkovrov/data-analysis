```python
#–ù–∞–≥–æ—Ä–Ω–æ–≤
dl = []
res = []
ch = 1500000
while len(res)<5:
    d = []
    total = 0
    for i in range(2,int(ch**0.5)+1):
        if ch%i==0:
            total += i
            total += ch//i
    flag = True
    for z in range(2,int(total**0.5)+1):
        if total % z == 0:
            flag=False
            break
    if flag and total != 0:
        res.append(ch)
        dl.append(total)
    ch+=1
print(*res)
print("--")
print(*dl)
```

```python
#–ù–∞–≥–æ—Ä–Ω–æ–≤
def f(x, y):
    if x == y:
        return 1

    if x < y or x == 13:
        return 0
    if x > y:
        return f(x - 1, y) + f(x//2, y) + f(x // 3, y)


print(f(26, 8) * f(8, 3))
```

# 16 –∑–∞–¥–∞–Ω–∏–µ

```python
#–ù–∞–≥–æ—Ä–Ω–æ–≤
from functools import lru_cache
@lru_cache(None)
def f(n):
    if n < 3:
        return 1
    if n > 2 and n%2 == 0:
        return f(n-1) + n - 1
    if n > 2 and n%2==1:
        return f(n-2)+2*n-2
for i in range(1,3048):
    f(i)
print(f(3048)-f(3045))
```
