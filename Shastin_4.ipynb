# задание 14

```python
#соловьев
s=12*(17**12)+13*(17**7)

def st(n):
    if n == 0:
        return [0]
    cifr = []
    while n:
        cifr.append(n % 17)
        n //= 17
    return cifr[::-1]

def cifr_above_9(digits):
    return sum(1 for d in digits if d > 9)

m=[]
for x in range(10000):
    N=s-x
    cifr17=st(N)
    if cifr_above_9(cifr17)==9:
       m.append(x)

print(max(m))
```

# задание 23

```python
# соловьев
from functools import lru_cache

@lru_cache(None)
def cp(n):
    if n == 12:
        return 1
    if n < 12:
        return 0
    total = 0
    # команда a: вычесть 2
    total += cp(n - 2)

    # команда b: поменять местами
    if n >= 10:
        units = n % 10
        tens = (n // 10) % 10
        if units < tens:
            swapped = units * 10 + tens
            total += cp(swapped)

    return total

result = cp(49)
print(result)
# может можно проще, но у меня не получалось
```

```python
# Пронин
from collections import Counter
with open('26.txt') as f:
    n=int(f.readline())
    # создаем словарь с ключом номера задачи
    # и значением множества учеников ее решивших
    d=dict()
    for i in range(n):
        id_student,id_problem = map(int,f.readline().split())
        if id_problem not in d:
            d[id_problem]=set()
        d[id_problem].add(id_student)
# выберем из словаря только те задачи, которые решил единственный человек
# и создадим список из номеров участников
rez=(list(d[i])[0] for i in d.keys() if len(d[i])==1)
# создадим словарь с ключом Id ученика и значением количества
# решенных задач только этим учеником
i=Counter(rez)
print(i)
# работает меньше секунды, но где-то ошибка, так как первая тройка ответов у меня такая
# 40645 117
# 42308 116
# 49362 103
# ошибка в несоответствии первого введенного числа и количества следующих строк
```
