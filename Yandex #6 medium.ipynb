# Задание 5

```python
def five(n):
    result=''
    while n>0:
        result=str(n%5)+result
        n=n//5
    return result
for n in range(1,10000):
    p=five(n)
    if n%25==0:
        p=p[-3:]+p
    else:
        p=p+five(n%25)
    r=int(p,5)
    if r>10000:
        print(n)
        break
```
