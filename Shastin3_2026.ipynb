```python
def f(n):
    s=''
    while n>0:
        s=str(n%2)+s
        n=n//2
    return s
for n in range(1,1000):
    a=f(n)
    if sum(map(int,a))%2==0:
        a=a+'0'
        a='10'+a[2:]
    else:
        a=a+'1'
        a='11'+a[2:]
    r=int(a,2)
    if n%2==1 and r%2==0 and r>367:
        print(n)
        break
```

```python
# Нагорнов
def prost(c):
    flag = True
    for x in range(2,int(c**0.5)+1):
        if c%x==0:
            flag = False
            break
    return flag

dl = []
otv = []
a=2142578
while len(otv) !=5:
    d=[]
    for i in range(1, int(a**0.5)+1):
        if a%i==0:
            if prost(i):
                if prost(a//i):
                    if (i+a//i) %2 != 0:
                        d.append(i)
                        d.append(a//i)
    if len(d) !=0:
        dl.append(max(d))
        otv.append(a)
    a+=1
print(*otv,sep="\n")
print("--")
print(*dl,sep="\n")
```

```python
#Пронин
def isPrime(n):
    for i in range(2,int(n**0.5)+1):
        if n%i==0:
            return False
    return True
def delit(n):
    for i in range(2,int(n**0.5)+1):
        if n%i==0:
            if isPrime(i) and isPrime(n//i) and (i+n//i)%2!=0:
                return (True,n//i)
    return (False,0)
k=0
for n in range(2142578,10**10):
    x,y=delit(n)
    if x==True:
        print(n,y)
        k+=1
        if k==5:
            break
```
